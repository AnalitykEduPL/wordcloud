```python
import imageio  # bundled with Anaconda
from pathlib import Path
from wordcloud import WordCloud
```

```python
mask_image = imageio.imread(r'mask_christmas.png')  # returns NumPy array of image's data
```

```python
wordcloud = WordCloud(width=1000, height=1000, 
    colormap='prism', mask=mask_image, background_color='white')
```

```python
text = Path(r'libraries.txt').read_text()
```

```python
wordcloud = wordcloud.generate(str(text))
```

```python
wordcloud = wordcloud.to_file('Libs_christmas.png')
```
